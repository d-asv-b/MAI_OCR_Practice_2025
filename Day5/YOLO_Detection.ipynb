Импорт библиотек

```python
from ultralytics import YOLO
from symspellpy import SymSpell, Verbosity
from google import genai

import os
import glob
import re
import json
import cv2
```

Константы:

```python
image_dir = "./images"
texts_dir = "./detected_text"
seg_result_dir = "./seg_results"
path_to_dictionary = "./ru_full.txt"
google_ai_api_key = "<google-genai-api-key>"
prompt_path = "./prompt.txt"

if not os.path.isdir(seg_result_dir):
    os.mkdir(seg_result_dir)

classes = [ "22", "23", "24", "25", "26", "27", "28" ]
```

Инициализация YOLO, Google GenAI и SymSpell

```python
yolo_model = YOLO("best.pt")
```

```python
gai_client = genai.Client(api_key=google_ai_api_key)
```

```python
sym_spell = SymSpell(max_dictionary_edit_distance=2, prefix_length=5)
with open(path_to_dictionary, "r", encoding="utf-8") as dict_file:
        if not sym_spell.create_dictionary(dict_file):
            raise Exception(f"Не удалось загрузить словарь из {path_to_dictionary}")
```

Фукнция очистки текста от лишних символов:

```python
def clear_text(text: str) -> str:
    return text.replace("\n", " ").replace("\r", " ").replace("\t", " ").replace("- ", "").strip()
```

Функция исправления грамматических ошибок в тексте по словарю с помощью библеотеки SymSpell:


```python
def fix_spelling_mistakes(text):
    clear = clear_text(text)
    scanned_words = clear.split(" ")
    converted_text = ""

    for word in scanned_words:
        prefix = ""
        suffix = ""
        core = word

        while core and not core[0].isalpha():
            prefix += core[0]
            core = core[1:]
        while core and not core[-1].isalpha():
            suffix = core[-1] + suffix
            core = core[:-1]

        clear_word = core.lower()
        if clear_word:
            suggestions = sym_spell.lookup(clear_word, Verbosity.CLOSEST, max_edit_distance=2)
            if suggestions:
                best_suggestion = suggestions[0].term

                if core.istitle():
                    best_suggestion = best_suggestion.capitalize()
                elif core.isupper():
                    best_suggestion = best_suggestion.upper()
                converted_text += f"{prefix}{best_suggestion}{suffix} "
            else:
                converted_text += f"{word} "
        else:
            converted_text += f"{word} "

    return converted_text
```

Функция нахождения текста, соответствующего определенном с помощью YOLO bounding box-у:

```python
def find_corressponding_text(work_id, image_num, bounding_box):
    detected_text_file_path = os.path.join(texts_dir, f"{work_id}_{image_num}__1_res.txt.webRes")

    text_in_box = ""
    with open(detected_text_file_path, "r") as json_file:
        text_data = json.load(json_file)


        for text_block in text_data["data"]["blocks"]:
            text_box = (text_block["x"], text_block["y"], text_block["w"], text_block["h"])

            ix, iy, iw, ih = text_box
            ox, oy, ow, oh = bounding_box

            # Calculate bottom-right corners
            inner_right = ix + iw
            inner_bottom = iy + ih
            outer_right = ox + ow
            outer_bottom = oy + oh

            # Tolerance in 10 px
            tolerance = 50
            if (
                ix >= ox - tolerance and iy >= oy - tolerance and
                inner_right <= outer_right + tolerance and
                inner_bottom <= outer_bottom + tolerance
            ):
                for box in text_block["boxes"]:
                    for lang in box["languages"]:
                        for text in lang["texts"]:
                            text_in_box = " ".join([text_in_box, text["text"]])

    return fix_spelling_mistakes(text_in_box)
```

Функция запуска предобученной YOLO модели для распознавания заданий на листах:

```python
def run_detection(work_id, image_num: str) -> dict:
    image_file_path = os.path.join(image_dir, f"{work_id}_{image_num}.png")
    results = yolo_model(image_file_path, conf=0.5, iou=0.4)

    save_path = os.path.join(seg_result_dir, f"work_{work_id}/result_{work_id}_{image_num}.png")
    # Saving image with boxes
    results[0].save(filename=save_path)

    boxes = []
    # Saving detection info for debug
    label_path = os.path.join(seg_result_dir, f"work_{work_id}/labels_{work_id}_{image_num}.txt")
    with open(label_path, 'w') as f:
        for box in results[0].boxes:
            cls = int(box.cls[0])
            conf = float(box.conf[0])
            xywh = box.xywh[0].tolist()
            f.write(f"{cls} {conf:.2f} {' '.join(map(str, xywh))}\n")

            boxes.append({ "ex": cls, "box": xywh })

    return boxes
```

Функция обработки изображения:

```python
def process_image(work_id: str, image_num: str) -> list:
    print(f"Processing image: {work_id}_{image_num}.png")

    bounding_boxes = run_detection(work_id, image_num)

    exercises_data = []
    for bb_data in bounding_boxes:
        box = bb_data["box"]
        x_center = box[0]
        y_center = box[1]
        width = box[2]
        height = box[3]

        x_tl = x_center - (width / 2)
        y_tl = y_center - (height / 2)

        text = find_corressponding_text(work_id, image_num, (x_tl, y_tl, width, height))
        exercises_data.append({"exerciseNumber": classes[bb_data['ex']], "text": text})

    print(f"\nDone processing image {work_id}_{image_num}.png!\n")
    return exercises_data
```

Функция обработки всей работы:

```python
def get_work_text_with_bounding_boxes(work_id, pages):
    result_json = []

    for page_num in pages:
        text_file_path = os.path.join(texts_dir, f"{work_id}_{page_num}__1_res.txt.webRes")
        page_text_with_bbs = {
            "pageNumber": page_num,
            "text": []
        }

        with open(text_file_path, "r") as json_file:
            text_data = json.load(json_file)

            for text_block in text_data["data"]["blocks"]:
                text_box = (text_block["x"], text_block["y"], text_block["w"], text_block["h"])

                ix, iy, iw, ih = text_box

                inner_right = ix + iw
                inner_bottom = iy + ih

                text_in_box = ""
                for box in text_block["boxes"]:
                    for lang in box["languages"]:
                        for text in lang["texts"]:
                            text_in_box = " ".join([text_in_box, text["text"]])

                page_text_with_bbs["text"].append({
                    "bbTopLeftX": ix,
                    "bbTopLeftY": iy,
                    "bbBottomRightX": inner_bottom,
                    "bbBottomRightY": inner_right,
                    "text": fix_spelling_mistakes(text_in_box)
                })

        result_json.append(page_text_with_bbs)

    with open(os.path.join(seg_result_dir, f"work_{work_id}", "text_with_bounding_boxes.json"), "w", encoding="utf-8") as result_file:
        json.dump(result_json, result_file, ensure_ascii=False)

    return result_json
```

```python
def process_work(work_id: str):
    print(f"Starting processing work {work_id}...\n\n")

    pattern = os.path.join(texts_dir, f"{work_id}_*__1_res.txt.webRes")
    matching_files = sorted(glob.glob(pattern))

    if not matching_files:
        print(f"No files found for work ID: {work_id}")
        return

    print(f"Found {len(matching_files)} file(s) for work ID {work_id}.")
    work_path = os.path.join(seg_result_dir, f"work_{work_id}")
    if not os.path.isdir(work_path):
        os.mkdir(work_path)

    page_ids = []
    for filepath in matching_files:
        filename = os.path.basename(filepath)
        match = re.match(rf"{re.escape(work_id)}_(\d+)_+.*", filename)
        if match:
            page_id = match.group(1)
            page_ids.append(page_id)
        else:
            print(f"Filename did not match expected pattern: {filename}")

    text_with_bounding_boxes = get_work_text_with_bounding_boxes(work_id, page_ids)

    work_data = []
    for page_id in page_ids:
        page_data = process_image(work_id, page_id)
        work_data.append({ "pageNumber": page_id, "data": page_data })

    with open(os.path.join(work_path, "work_data.json"), "w", encoding="utf-8") as result:
        json.dump(work_data, result, ensure_ascii=False)

    print(f"\nDone processing work {work_id}!\n\n")
    return work_data
```

Подключение LLM для оценки заданий:

```python
with open(prompt_path, "r", encoding="utf-8") as prompt_file:
    prompt = "".join(prompt_file.readlines())
```

```python
def evaluate_work(work_data):
    exercises = ""
    already_seen = []
    for page in work_data:
        for exercise in page["data"]:
            ex_number = exercise["exerciseNumber"]
            if ex_number != 24 and ex_number != 28 and not ex_number in already_seen:
                exercises += f"\nЗадание {ex_number}: {exercise['text']}"
                already_seen.append(ex_number)

    result = gai_client.models.generate_content(
        model="gemini-2.0-flash",
        contents=prompt + exercises
    )

    pattern = r"(?:json)?\s*(\[\s*{.*?}\s*])\s*"
    match = re.search(pattern, result.text, re.DOTALL)

    if match:
        json_str = match.group(1)
        data = json.loads(json_str)
        return data
    else:
        print("JSON not found.")

    return ""
```

```python
res = process_work("1019641987")
```
